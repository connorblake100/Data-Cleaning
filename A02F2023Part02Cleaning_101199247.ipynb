### Step 2d. (14 pts) Visually inspect the box plots and determine if there are any statistical outliers for each of the two attributes. Remember, a statistical outlier is any data point that exceeds the upper and lower whiskers. For each attribute that contains outliers:
1) Calculate and display the IQR
2) Calculate and display the necessary min and/or max that are the thresholds to determine outliers (use the <span style="color:blue">75th + 1.5 IQR</span> and <span style="color:blue">25th - 1.5 IQR</span> formula discussed in class)
3) Use the above information to display the row(s) that contain outliers

```python
# function to detect outliers
def outlier_finder(column_name):
    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - (1.5*IQR)
    upper = Q3+(1.5*IQR)
    outliers = df[(df[column_name] < lower) | (df[column_name] > upper)]
    print(outliers)
```

```python
# outlier for unemployment_rate_female
outlier_finder('unemployment_rate_female')
```

```python
# outlier for unemployment_rate_male
outlier_finder('unemployment_rate_male')
```
